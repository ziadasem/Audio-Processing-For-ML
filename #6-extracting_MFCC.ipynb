```python
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
```

```python
audio_file = "audio/debussy.wav"
signal, sr = librosa.load(audio_file)
```

```python
mfcc = librosa.feature.mfcc(signal, n_mfcc=13, sr=sr)
```

```python
mfcc.shape
```

```python
plt.figure(figsize=(25,10))
librosa.display.specshow(mfcc, sr=sr ,x_axis="time" )
plt.colorbar(format="%+2f")
plt.show()
```

```python
delta_mffc = librosa.feature.delta(mfcc)
delta_mffc2 =  librosa.feature.delta(mfcc, order=2)
```

```python
delta_mffc2.shape
```

```python
plt.figure(figsize=(25,10))
librosa.display.specshow(delta_mffc, sr=sr ,x_axis="time" )
plt.colorbar(format="%+2f")
plt.show()
```

```python
plt.figure(figsize=(25,10))
librosa.display.specshow(delta_mffc2, sr=sr ,x_axis="time" )
plt.colorbar(format="%+2f")
plt.show()
```
