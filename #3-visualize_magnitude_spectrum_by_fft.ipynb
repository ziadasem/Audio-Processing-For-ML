```python
#convert these sound to numpy array
violin_c4 , sr = librosa.load(violin_file)
sax_c4 , sr = librosa.load(sax_file)
piano_c5 , sr = librosa.load(piano_file)
noise , sr = librosa.load(noise_file)
```

```python
#plot mgnitude spectrum
def plot_magnitude_spectrum(signal, title, sr, f_ratio = 1):
  ft = np.fft.fft(signal)
  magnitude_spectrum =np.abs(ft)

  plt.figure(figsize=(18,5))
  frequency = np.linspace(0, sr, len(magnitude_spectrum))
  num_frequency_bins = int(len(frequency)* f_ratio)
  plt.plot(frequency[:num_frequency_bins], magnitude_spectrum[:num_frequency_bins])
  plt.xlabel("Frequency in Hz")
  plt.title(title)

  plt.show()
```

```python
plot_magnitude_spectrum(violin_c4, "violin", sr, 0.1)
```

```python
plot_magnitude_spectrum(sax_c4, "sax", sr, 0.1)
```

```python
plot_magnitude_spectrum(piano_c5, "piano", sr, 0.1)
```

```python
plot_magnitude_spectrum(noise, "noise", sr, 0.1)
```
